## Importing necessary libraries

```python
from sklearn.linear_model import SGDRegressor
from sklearn.datasets import load_boston
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
```

### Data preparation

```python
x, y = make_regression(n_samples = 5000, n_features = 30)
```

```python
x.shape
```

```python
y.shape
```

```python
x = scale(x)
y = scale(y)
```

```python
x_train,x_test,y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
sgd_reg = SGDRegressor()
```

```python
m1 = sgd_reg.fit(x_train,y_train)
```

```python
score = sgd_reg.score(x_train, y_train)
```

```python
print("R-Squared: ", score)
```

### Cross Validation

```python
cv_score = cross_val_score(sgd_reg,x, y, cv = 10)
```

```python
print("cross valization score: ", cv_score.mean())
```

### Predicting 

```python
y_pred = m1.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print("Mean Squared Error: ", mse)
print("Root Mean Squre Error: ", mse**1/2)
```

### Visualizing actual vs predicted

```python
x_axis = range(len(y_test))
plt.plot(x_axis, y_test, linewidth =1, label = 'original')
plt.plot(x_axis, y_pred, linewidth =1, label = 'predicted')
plt.title( 'Actual vs Predicted')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.legend(loc = 'best', fancybox = True, shadow = True)
plt.grid(True)
plt.show()
```

## Using Boston dataset

```python
boston = load_boston()
x, y = boston.data, boston.target
```

```python
x.shape
```

```python
y.shape
```

```python
x = scale(x)
y = scale(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
sgd_reg = SGDRegressor(alpha = 0.0001, epsilon = 0.01, eta0 = 0.1, penalty = 'elasticnet')
```

```python
m2 = sgd_reg.fit(x_train, y_train)
```

```python
y_pred = m2.predict(x_test)
```

```python
error = mean_squared_error(y_test, y_pred)
```

```python
print("mean squared error: ", error)
print("root mean squared error: ", error**1/2)
```

### Cross Validation

```python
cv_score = cross_val_score(sgd_reg, x, y, cv = 10)
```

```python
print("cross validation score: ", cv_score.mean())
```

```python
x_axis = range(len(y_test))
plt.plot(x_axis, y_test, linewidth =1 , label = 'original')
plt.plot(x_axis, y_pred, linewidth =1 , label = 'predicted')
plt.title('Actual vs Predicted')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend(loc = 'best', fancybox = True, shadow = True)
plt.grid(True)
plt.show()
```
